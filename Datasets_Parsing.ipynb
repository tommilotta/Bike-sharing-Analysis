# Data Parsing

First homogenize Chicago dataset with the other two

```python
CHI_df.drop(columns=['01 - Rental Details Rental ID'], inplace=True)

# renaming colmuns
CHI_df.set_axis(['starttime', 'stoptime', 'bikeid', 'tripduration', 'start station id', 'start station name', 
                  'end station id', 'end station name', 'usertype', 'gender', 'birth year'], axis=1, inplace=True)

# changing Male and Female with 1 and 2 (nan = 0)
CHI_df = CHI_df.astype({'gender': str})
CHI_df['gender'] = CHI_df['gender'].str.replace('Male', '1')
CHI_df['gender'] = CHI_df['gender'].str.replace('Female', '2')
CHI_df['gender'] = CHI_df['gender'].str.replace('nan', '0')
CHI_df = CHI_df.astype({'gender': int})

# removing commas from tripduration
CHI_df['tripduration'] = CHI_df['tripduration'].str.replace(',', '')
```

Below start hour and week day columns are added from starttime. (Remember : Chicago's dataset has also other months other than April so we have to drop some data)

```python
from datetime import datetime as dt

# Extracting starting, ending hour and day of the week (Monday = 0, Tuesday = 1, .., Sunday = 6)
def extract_hour(df):
    datetimes = pd.to_datetime(df['starttime'])
    df['starttime'] = datetimes
    df['start hour'] = df['starttime'].dt.hour
    df['week day'] = df['starttime'].dt.dayofweek
    df = df[df['starttime'].dt.month == 4]
    return df

NY_df = extract_hour(NY_df)
CHI_df = extract_hour(CHI_df)
BOS_df = extract_hour(BOS_df)
```

Divvy's dataset for the chosen year is missing data about stations latitude and longitude.

In order to get them the dataset from 2020 is used.

```python
chi_2020 = pd.read_csv('202004-divvy-tripdata.zip', compression='zip', header=0, sep=',', quotechar='"', low_memory=False)

# creation of the lat-long df
chi_lat_long = chi_2020[['start_station_name', 'start_lat', 'start_lng']]
chi_lat_long.set_axis(['end_station_name', 'end_lat', 'end_lng'], axis=1, inplace=True)
chi_lat_long = chi_lat_long.append(chi_2020[['end_station_name', 'end_lat', 'end_lng']])
chi_lat_long.drop_duplicates(inplace=True, ignore_index=True)

# merging with original dataframe
CHI_df = pd.merge(left=CHI_df, right=chi_lat_long, how='left', left_on='start station name', right_on='end_station_name')
CHI_df = pd.merge(left=CHI_df, right=chi_lat_long, how='left', left_on='end station name', right_on='end_station_name')
CHI_df.rename(columns={'end_lat_x' : 'start station latitude', 'end_lng_x' : 'start station longitude',
               'end_lat_y' : 'end station latitude', 'end_lng_y' : 'end station longitude'}, inplace=True)

# dropping rows with null values for latitude and longitude
CHI_df.dropna(subset=['start station latitude', 'start station longitude', 'end station latitude',
                      'end station longitude'], inplace=True)

# reordering columns
CHI_df = CHI_df[list(NY_df.columns)]
CHI_df.head()
```

### Saving the three parsed dataframes as csv files

```python
NY_df.to_csv('data/newyork_dataset', index=False)
CHI_df.to_csv('data/chicago_dataset', index=False)
BOS_df.to_csv('data/boston_dataset', index=False)
```
